```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv("Downloads/basket.csv")
```

```python
df.head(30)
```

```python
cols = list(df.columns)
cols
```

```python
for col in cols:
    df[col] = df[col].fillna('')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
one_hot_encoded = pd.get_dummies(df)
```

```python
one_hot_encoded
```

```python
frequent_sets = apriori(one_hot_encoded, min_support=0.4, use_colnames=True)
frequent_sets.head(30)
```

```python
frequent_sets.shape
```

```python
association_df = association_rules(frequent_sets, min_threshold=0.7, metric="confidence")
association_df
```

```python
association_df.head(30)
```

```python
association_df.tail(30)
```

```python
association_df.shape
```
